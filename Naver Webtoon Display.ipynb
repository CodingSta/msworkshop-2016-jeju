```python
import requests
from bs4 import BeautifulSoup
from IPython.display import display, Image


def main():
    ep_url = 'http://comic.naver.com/webtoon/detail.nhn?titleId=675554&no=100&weekday=mon'
    html = requests.get(ep_url).text
    soup = BeautifulSoup(html, 'html.parser')
    for img_tag in soup.select('.wt_viewer img'):
        img_url = img_tag['src']
        img_data = requests.get(img_url, headers={'Referer': ep_url}).content
        print(img_url)
        display(Image(img_data))


if __name__ == '__main__':
    main()
```
